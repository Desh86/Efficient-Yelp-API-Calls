```python
from yelpapi import YelpAPI
import json
import pandas as pd
```

```python
with open('/Users/user/.secret/yelp_api.json') as f:
    login = json.load(f)
login.keys()
```

```python
yelp_api = YelpAPI(login['api-key'], timeout_s=5.0)
yelp_api
```

```python
results = yelp_api.search_query(location='Greenville, SC',
                                       term='pizza')
print(type(results))
results.keys()
```

```python
results1 = pd.DataFrame(results['businesses'])
results1.head(2)
```

```python
results['total']
```

```python
results_all = yelp_api.search_query(location='Greenville, SC',
                                       term='pizza',
                               offset = 0)
```

```python
resultsall = pd.DataFrame(results_all['businesses'])
resultsall.head(5)
```

```python
#now concatenate all pages
Concat_results = pd.concat([results1, resultsall],
                      ignore_index=True)
display(Concat_results.head(3), Concat_results.tail(3))
```

```python
Concat_results.to_json(orient='records')
```
